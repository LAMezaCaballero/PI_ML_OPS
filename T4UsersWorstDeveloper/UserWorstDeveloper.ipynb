 |Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

```python
#caargar los datos
import pandas as pd
import os
```

```python
#aprovecho que ya tengo el analisis de sentimientos del proceso anterior
d_Reviews = pd.read_parquet('UsersRecommendNeg.parquet')
```

```python
#ahora solo necesito juntar el Item_id con el developer.
D_dev= pd.read_json(r'..\dataset_raw\Clean_DataGames.json')
D_dev= D_dev[['item_id', 'developer']].drop_duplicates()
D_dev
```

```python
# las mezclamos y eliminamos columnas innecesarias
d_master= d_Reviews.merge(D_dev, how='inner', on='item_id')
d_master.drop(columns={	'item_id',	'item_name'}, axis= 1,inplace=True)

```

```python
d_master
```

```python
#crear la lista de solo los 3 peores por año
top_3_per_year = d_master.groupby('release_date').head(3)# se obtiene las primeras 3 para cada ano y se almacenan
top_3_per_year.sort_values(by=['release_date', 'sentiment_analysis'],ascending=True, inplace=True)
top_3_per_year
```

```python
top_3_per_year.drop(columns={'sentiment_analysis'}, inplace=True)
top_3_per_year.head(2)
```

```python
top_3_per_year.to_parquet('UsersRecommenT4.parquet')
```

```python
def UsersWorstDeveloper( año : int ): 
    ######verificar que este el año pedido ente los datos
    if año not in top_3_per_year['release_date'].unique() or type(año) != int :
        raise ValueError(f"El '{año}' no se encuentra en la base de datos favor de verificarlo, puede que no sea un numero.")
    
    ## extraer los datos del ano
    df_ano= top_3_per_year[top_3_per_year['release_date']==año]
    
    ##revisar que no sean vacios.
    nueva_fila = pd.DataFrame([{'release_date': año, 'developer': 'no hay'}])
    
    if df_ano.size ==9:
        return print([{"Puesto 1" : df_ano['developer'].iloc[0]}, {"Puesto 2" : df_ano['developer'].iloc[1]},{"Puesto 3" : df_ano['developer'].iloc[2]}])  
    
    elif df_ano.size <4:
       
        df_ano = pd.concat([df_ano, nueva_fila], ignore_index=True)
        df_ano = pd.concat([df_ano, nueva_fila], ignore_index=True)  
        return print([{"Puesto 1" : df_ano['developer'].iloc[0]}, {"Puesto 2" : df_ano['developer'].iloc[1]},{"Puesto 3" : df_ano['developer'].iloc[2]}])   
      
    else :
        df_ano = pd.concat([df_ano, nueva_fila], ignore_index=True)  
        return print([{"Puesto 1" : df_ano['developer'].iloc[0]}, {"Puesto 2" : df_ano['developer'].iloc[1]},{"Puesto 3" : df_ano['developer'].iloc[2]}])  
    
    #return  df_ano
    #return print([{"Puesto 1" : df_ano['developer'].iloc[0]}, {"Puesto 2" : df_ano['developer'].iloc[1]},{"Puesto 3" : df_ano['developer'].iloc[2]}])

```
