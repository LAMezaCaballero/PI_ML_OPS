Si es un sistema de recomendación item-item:

### def recomendacion_juego( id de producto ): 
    Ingresando el id de producto, deberíamos recibir una lista con 5 juegos recomendados similares al ingresado.


crear la base de datos

```python
#cargar librerias
import os
import pandas as pd 
```

```python
#cargar datos del juego, 
# Clean_DataGames_T3 me da el nombre del juego, el item id.
data= pd.read_json(r'../etl/dataset_raw/Clean_DataGames_T3.json')
data.head(5)
```

```python
data.drop(columns={'release_date'},axis=1,inplace=True)# eliminar la columna release_date
data.dropna(subset=['item_id'], inplace=True) # eliminar los vacios de item_id
data['item_id']=data['item_id'].astype(int) # cabiar a tipo int el valor de item_id
data.head(5)
```

```python
#Clean_DataGames me da el genero del juego, el item id.
data2= pd.read_json(r'../etl/dataset_raw/Clean_DataGames.json')
data2.head(5)
data2[data2['item_id']==761140]
```

```python
data2.drop(columns={'release_date','developer'},axis=1,inplace=True)# eliminar la columna release_date y developer
data2.dropna(subset=['item_id'], inplace=True) # eliminar los vacios de item_id

data2.head(5)
```

```python
df= pd.merge(data2, data, on='item_id')
df
```

```python
genres_dummies = pd.get_dummies(df['genres'])
genres_dummies
```

```python
df = pd.concat([df, genres_dummies], axis=1)
```

```python
df.drop(columns=['genres'], inplace=True)
```

```python
df
```

```python
#reducimos las clasificaciones a las mas importantes.
df.drop(columns={'ACCOUNTING','SPORTS', 'ANIMATION &AMP; MODELING','AUDIO PRODUCTION',	'CASUAL',	'DESIGN &AMP; ILLUSTRATION' ,'EARLY ACCESS','SOFTWARE TRAINING','FREE TO PLAY','PHOTO EDITING','UTILITIES','VIDEO PRODUCTION','WEB PUBLISHING'},axis=1,inplace=True)
df
```

```python
grouped_data = df.groupby(['item_id', 'item_name']).max().reset_index()
grouped_data.head()
```

```python
grouped_data[grouped_data['item_id']==761140]
```

```python
#guardar el df para el ML
grouped_data.to_parquet('recomendacion_juego.parquet',index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Obtener las características de los ítems (columnas desde 'ACTION' en adelante)
features = grouped_data.iloc[:, 2:]

# Calcular la similitud de coseno entre los vectores de características
cosine_sim = cosine_similarity(features, features)

# Obtener el índice del ítem por el que quieres encontrar similitudes
target_index = 0  # Cambia este valor según el índice del ítem que deseas buscar similitudes

# Ordenar los ítems según la similitud de coseno y obtener los 5 ítems más similares
similar_items_indices = cosine_sim[target_index].argsort()[-6:-1][::-1]

# Obtener los nombres de los ítems similares
similar_items_names = data.iloc[similar_items_indices]['item_name'].values

# Imprimir los nombres de los 5 ítems más similares
print("Los 5 ítems más similares a", data.at[target_index, 'item_name'], "son:")
for name in similar_items_names:
    print(name)
```

```python
#2 veamos para el 10, 20, 2028850
from sklearn.metrics.pairwise import cosine_similarity

def recomendacion_juego( id_de_producto : int): 
        
    # Filtrar el DataFrame para obtener solo las filas correspondientes al item_id de consulta
    consulta = grouped_data[grouped_data['item_id'] == id_de_producto].iloc[:, 2:]

    # Calcular la similitud de coseno entre el item de consulta y todos los demás items
    similaridades = cosine_similarity(consulta, grouped_data.iloc[:, 2:])

    # Obtener los índices de los items más similares (excepto el item de consulta)
    indices_mas_similares = similaridades.argsort(axis=1)[0][-6:-1]

    # Obtener los nombres de los items más similares
    nombres_mas_similares = grouped_data.loc[indices_mas_similares, 'item_name']

    llave= f"Los 5 item_name más similares al item_id {item_id_consulta}son:"
    # Mostrar los nombres de los items más similares
    return {llave : nombres_mas_similares}
```
