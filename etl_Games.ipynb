```python
import pandas as pd
import numpy as np
import os
import json
from pandas import json_normalize
import ast
```

cargar los datos del archivo output_steam_games.json

```python

ruta_archivo_json = r'output_steam_games.json'

# Lee el archivo JSON y carga los datos en una lista de diccionarios
data_list = []
with open(ruta_archivo_json, 'r', encoding='utf-8') as file:
    for line in file:
        
        data_list.append(json.loads(line.strip()))#carga la linea y limpia espacios vacios 

a= pd.DataFrame(data_list) # lo convierte en DF
```

```python
a.head()
```

```python
#se elimina filas completas por vacio
a.dropna(how='all',inplace=True)
a.reset_index(drop=True)
```

brincar para T3UserRecommends para trabajar

```python
# se eliminan las columnas inncesarias para las peticiones
a.drop(['publisher','title','url','tags','reviews_url','specs','price','early_access','app_name'],axis=1, inplace=True)
a
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
#al revisar con info() se observo que hay muchos espacios vacios
a = a.dropna(subset=['genres','release_date','id','developer'])
a.reset_index(drop=True, inplace=True)
a
```

```python
a.isnull().sum()
```

```python
# Convertir la columna 'release_date' al tipo datetime, ignorando valores no válidos
a['release_date'] = pd.to_datetime(a['release_date'], errors='coerce')

# Extraer el año y asignarlo nuevamente a la columna 'release_date'
a['release_date'] = a['release_date'].dt.year
```

```python
#expandir los generos para que tengan cada uno su fila
Genlist = a.explode('genres')
Genlist
```

```python
#Genlist = Genlist[['release_date', 'genres', 'id']]
Genlist.reset_index(drop=True, inplace=True)
```

```python
Genlist
```

#cambiar los datos al tipo correcto release_date as int, genres a mayusculas, y el nombre de la columna id por item_id  

```python
Genlist['genres'] = Genlist['genres'].str.upper()
Genlist['developer'] = Genlist['developer'].str.upper()
Genlist.dropna(subset=['release_date'], inplace=True)  # Drop rows with NaN in 'release_date'
Genlist['release_date'] = Genlist['release_date'].astype('int')
Genlist=Genlist.rename(columns={'id':'item_id'})
Genlist
```

Datos limpios guardados para su uso.

```python
Genlist.to_json('Clean_DataGames.json',index=False)
```
