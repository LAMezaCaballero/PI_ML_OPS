```python
def analyze_sentiment(review):
  '''
  Retorno:
  1 si el sentimiento es positivo, -1 si es negativo, o 0 si es neutral.
  '''

  # Define palabras positivas y negativas (puedes ajustar estas listas)
  positive_words = ["good", "great", "amazing", "fun", "love", "recommend", ':)', '♥','beast ','best','amazing' ]
  negative_words = ["bad", "terrible", "boring", "hate", "disappointed", "not",'sucks ','fix','repetitive','un-stable','non-structured' ]

  # Inicializa el conteo de sentimiento
  sentiment = 0

  # Separa las palabras de la reseña
  words = review.lower().split()

  # Itera por cada palabra
  for word in words:
    if word in positive_words:
      sentiment += 1
    elif word in negative_words:
      sentiment -= 1

  return sentiment


```

```python
  # Agrega la columna 'sentiment_analysis'
merged['sentiment_analysis'] = merged['review'].apply(analyze_sentiment)

merged
```

```python
from langdetect import detect

def analyze_language(review):
  """
  Analiza el idioma de un texto.

  Parámetros:
    review: El texto que se quiere analizar.

  Retorno:
    El código del idioma detectado o 'unknown' si no se puede determinar.
  """

  try:
    # Intenta detectar el idioma
    language = detect(review)
  except:
    language = 'unknown'

  return language

def recommend_sentiment_analysis(row):#-----------------------------agregar 1 si recommend == true
  """
  Actualiza el valor de sentiment_analysis basado en la recomendación (recommend).

  Parámetros:
    row: Una fila del DataFrame.

  Retorno:
    El valor actualizado de sentiment_analysis.
  """

  sentiment = row['sentiment_analysis']
  recommend = row['recommend']

  if recommend:
    sentiment += 1
  else:
    sentiment -= 1

  return sentiment

```

```python
merged['language'] = merged['review'].apply(analyze_language)
merged['sentiment_analysis'] = merged.apply(recommend_sentiment_analysis, axis=1)
merged
```

```python
def UsersRecommend( año : int ): 
    ######verificar que este el año pedido ente los datos
    if año not in df_Sentiment['release_date'].unique() or type(año) != int :
        raise ValueError(f"El género '{año}' no se encuentra en la base de datos favor de verificarlo.")
    
    ## extraer los datos del ano
    df_ano= df_Sentiment[df_Sentiment['release_date']==año]

    return print([{"Puesto 1" : df_ano['item_name'].iloc[0]}, {"Puesto 2" : df_ano['item_name'].iloc[1]},{"Puesto 3" : df_ano['item_name'].iloc[2]}])
```
