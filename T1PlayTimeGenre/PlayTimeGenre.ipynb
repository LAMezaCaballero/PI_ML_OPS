# Crear la funcion para la endpoint

    def PlayTimeGenre( genero : str ): Debe devolver año con mas horas jugadas para dicho género.

    Ejemplo de retorno: {"Año de lanzamiento con más horas jugadas para Género X" : 2013}

```python
#cargar el archivo
import os
import pandas as pd
import numpy as np
```

```python
df_Gen = pd.read_parquet('Data_playtimegender.parquet')
df_Gen
```

```python
# la funcion

def PlayTimeGenre( genero : str ):
    genero = genero.upper() #para evitar errores de mayusculas, todo se hizo mayusculas

    #revisar que el genero si sea un str y que este dentro del listado de generos posibles
    if genero not in df_Gen['genres'].unique():
        raise ValueError(f"El género '{genre}' no se encuentra en la base de datos favor de verificarlo.")
    
   
    df_query = df_Gen[df_Gen['genres'] == genero] #se extrae la parte del genero elegido

    # Encontrar el año con el mayor valor acumulado de playtime_forever
    df_query=df_query.sort_values(by='playtime_forever', ascending=False)

    return print ('Año de lanzamiento con más horas jugadas para Género ' ,genero,' : ', df_query['release_date'].iloc[0])
```
